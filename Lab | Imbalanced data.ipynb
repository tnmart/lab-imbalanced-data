#### 1. Load the dataset and explore the variables.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading dataset
df = pd.read_csv('files_for_lab/customer_churn.csv')
display(df.head())
display(df.tail())
```

```python
#checking for number of unique values
df.nunique()
```

```python
#checking for missing values
df.isna().sum()
```

```python
df.dtypes
```

```python
#the column senior citizen seems to be categorical where 1 corresponds to true and 0 to False. I will then replace
#the values accordingly and change the data type of the column
df['SeniorCitizen'].replace({1: "True", 0: "False"}, inplace=True)

display(df.head())
display(df.tail())
```

2. We will try to predict variable Churn using a logistic regression on variables tenure, SeniorCitizen,MonthlyCharges.
3. Split the Dataset into X ('tenure', 'SeniorCitizen', 'MonthlyCharges') and y ('Churn')

```python
#X/y Split
X = df[['tenure', 'SeniorCitizen', 'MonthlyCharges']]
y = df['Churn']
```

```python
#checking the correlation matrix for multicollinearity
correlations_matrix = X.corr()
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

There doesn't seem to exist multicollinearity between the X columns

```python
#Train/test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
#Splitting X into categorical and numerical columns
X_train_num = X_train.select_dtypes(include = np.number)
X_test_num  = X_test.select_dtypes(include = np.number)

X_train_cat = X_train.select_dtypes('object')
X_test_cat  = X_test.select_dtypes('object')

display(X_train_num)
display(X_test_num)
display(X_train_cat)
display(X_test_cat)
```

```python
#applying MinMaxScaler to the numerical columns
from sklearn.preprocessing import MinMaxScaler

#fitting the train set
MinMaxScaler = MinMaxScaler()
transformer = MinMaxScaler.fit(X_train_num)

#Saving scaler
import pickle
import os

path = "scalers/"

isExist = os.path.exists(path)
if not isExist:
  os.makedirs(path)
  print("The new directory is created!")

filename = "MinMaxScaler.pkl"
with open(path+filename, "wb") as file:
    pickle.dump(MinMaxScaler, file)
```

```python
#transforming the train and test sets
X_train_normalized = transformer.transform(X_train_num)
X_test_normalized  = transformer.transform(X_test_num)

#transforming the arrays into dataframes
X_train_norm = pd.DataFrame(X_train_normalized, columns=X_train_num.columns, index= X_train_num.index)
X_test_norm  = pd.DataFrame(X_test_normalized, columns=X_test_num.columns, index= X_test_num.index)
```

```python
display(X_train_norm.shape)
display(X_test_norm.shape)
X_train_norm.describe().T
```

```python
# applying OneHotEncoder to the SeniorCitizen column
from sklearn.preprocessing import OneHotEncoder

#fitting on the train set
encoder = OneHotEncoder(drop='first')
encoder.fit(X_train_cat)

#saving encoder
path = "encoders/"

isExist = os.path.exists(path)
if not isExist:
  os.makedirs(path)
  print("The new directory is created!")

filename = "OneHotEncoder.pkl"
with open(path+filename, "wb") as file:
   pickle.dump(encoder, file)
```

```python
#transforming both the train and test
X_train_cat_np = encoder.transform(X_train_cat).toarray()
X_test_cat_np  = encoder.transform(X_test_cat).toarray()

#converting to dataframes
X_train_cat = pd.DataFrame(X_train_cat_np, columns=encoder.get_feature_names_out(), 
                           index=X_train_cat.index)
X_test_cat = pd.DataFrame(X_test_cat_np,  columns=encoder.get_feature_names_out(),
                          index=X_test_cat.index)
```

```python
display(X_train_cat)
display(X_test_cat)
```

```python
#building X_train
X_train_transformed = pd.concat([X_train_norm, X_train_cat], axis=1)
X_train_transformed
```

```python
#building X_test
X_test_transformed = pd.concat([X_test_norm, X_test_cat], axis=1)
X_test_transformed
```

#### 4. Build the logistic regression model.

```python
#creating logistic regression model
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial')

lr.fit(X_train_transformed, y_train)

#saving model
path = "models/"
    
isExist = os.path.exists(path)
if not isExist:
    os.makedirs(path)
    print("The new directory is created!")

filename = "LogisticRegression.pkl" 
with open(path+filename, "wb") as file:
    pickle.dump(lr, file)
```

#### 5. Evaluate the model.

```python
#evaluating the model using score metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, cohen_kappa_score

y_pred_test = lr.predict(X_test_transformed)

print("The accuracy in the TEST set is: {:.2f}".format(accuracy_score(y_test,y_pred_test)))
print("The precision in the TEST set is: {:.2f}".format(precision_score(y_test,y_pred_test, pos_label='Yes')))
print("The recall in the TEST set is: {:.2f}".format(recall_score(y_test,y_pred_test, pos_label='Yes')))
print("The F1 in the TEST set is: {:.2f}".format(f1_score(y_test,y_pred_test, pos_label='Yes')))
print("The Kappa in the TEST set is: {:.2f}".format(cohen_kappa_score(y_test,y_pred_test)))
```

```python
#checking the Confusion Matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm_test = confusion_matrix(y_test,y_pred_test)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=lr.classes_)
disp.plot()
plt.show()
```

#### 6. Even a simple model will give us more than 70% accuracy. Why?

```python
count_churn = df['Churn'].value_counts()
print(count_churn)
sns.countplot(x = 'Churn', data = df);
plt.show()
```

Answer: Since there is a really big class imbalance, the model can have a really high accuracy by simply predicting the majority class most of the time which is what happens in this case.

#### 7. Synthetic Minority Oversampling TEchnique (SMOTE) is an over sampling technique based on nearest neighbors that adds new points between existing points. Apply imblearn.over_sampling.SMOTE to the dataset. Build and evaluate the logistic regression model. Is it there any improvement?

```python
#applying SMOTE
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=100,k_neighbors=3)

X_train_SMOTE,y_train_SMOTE = sm.fit_resample(X_train_transformed,y_train)
```

```python
print(X_train.shape)
print(X_train_SMOTE.shape)
```

```python
y_train_SMOTE.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_SMOTE, y_train_SMOTE)
y_pred_test = lr.predict(X_test_transformed)

print("The accuracy in the TEST set is: {:.2f}".format(accuracy_score(y_test,y_pred_test)))
print("The precision in the TEST set is: {:.2f}".format(precision_score(y_test,y_pred_test, pos_label='Yes')))
print("The recall in the TEST set is: {:.2f}".format(recall_score(y_test,y_pred_test, pos_label='Yes')))
print("The F1 in the TEST set is: {:.2f}".format(f1_score(y_test,y_pred_test, pos_label='Yes')))
print("The Kappa in the TEST set is: {:.2f}".format(cohen_kappa_score(y_test,y_pred_test)))

The accuracy in the TEST set is: 0.78
The precision in the TEST set is: 0.61
The recall in the TEST set is: 0.46
The F1 in the TEST set is: 0.52
The Kappa in the TEST set is: 0.39
```

```python
cm_test = confusion_matrix(y_test,y_pred_test)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=lr.classes_)
disp.plot()
plt.show()

```

Answer: Using the SMOTE technique we find the following relative to the first logistic regression model:
- moderate decrease in accuracy and precision scores;
- very small decrease in kappa score;
- moderate increase in F1 score;
- a relatively high increase in  recall score;

The high increase in recall score is possibly due to the fact that the model is better at predicting the minority samples from y while the decrease in accuracy shows us that is not so good in predicting the majority sample.
